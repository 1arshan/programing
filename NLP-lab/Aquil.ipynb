```python
import csv
import math
file = open('story.text', 'r')
text = file.read()
```

```python
words = text.split(' ')
print('total words:', len(words))
print('words:')
for word in words:
    print(word)
```

```python
print('total vocabulary: ',len(set(words)))
```

```python
sentences = text.split('.')
print('total sentence: ', len(sentences))
print('sentences: ')
for s in sentences:
    print(s+".")
```

```python
sentences = set((sentences))
sentences = tuple(sentences)
print('unique sentences:')
for i in range(len(sentences)):
    print(i, sentences[i])
```
