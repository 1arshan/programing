```python
words = text.strip().split(" ")
print(f"Vocabulary size: {len(words)}")
words
```

```python
import collections
dictionaryFrequency = collections.defaultdict(int)

for word in words:
    dictionaryFrequency[' '.join(word) + ' </w>'] += 1

dictionaryFrequency 
```

```python

countingChars = collections.defaultdict(int)
for word, freq in dictionaryFrequency.items():
    chars = word.split()
    for char in chars:
        countingChars[char] += freq

countingChars
```

```python
import re
# gets the pairs frequencies of chars
def PAIRS(dictionaryFrequency):
    freqpairs = collections.defaultdict(int)
    for word, freq in dictionaryFrequency.items():
        chars = word.split()
        for i in range(len(chars)-1):
            freqpairs[chars[i], chars[i+1]] += freq
    return freqpairs

# changes best pair to single char with no space in between them
def MBP(best_pair, dictionaryFrequency):
    # print(best_pair)
    merged_dict = {}
    gram = re.escape(' '.join(best_pair))
    p = re.compile(r'(?<!\S)' + gram + r'(?!\S)')
    for word in dictionaryFrequency:
        # print(word)
        w_out = p.sub(''.join(best_pair), word)
        merged_dict[w_out] = dictionaryFrequency[word]
    return merged_dict

def subwords(dictionaryFrequency):
    countingChars = collections.defaultdict(int)
    for word, freq in dictionaryFrequency.items():
        chars = word.split()
        for char in chars:
            countingChars[char] += freq
    return countingChars

for i in range(500):
    pairs = PAIRS(dictionaryFrequency)
    print(pairs)
    if (len(pairs) == 0):
      print("*********")
      print("Sentence is completely tokenized. No more words to merge")
      break

    best_pair = max(pairs, key=pairs.get) #max freq of pair
    # print("hello",best_pair)
    # print(f"Iteration {i}: ")
    dictionaryFrequency = MBP(best_pair, dictionaryFrequency)
    # print(temp)
    subword_tokens = subwords(dictionaryFrequency)
    print(subword_tokens)
    print("the no. of tokens is = {0}".format(len(subword_tokens)))
    print("*********")

```
