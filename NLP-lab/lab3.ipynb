```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
text = open('1.txt', 'r').read() # read a text file
sentences = sent_tokenize(text)
    
```

```python
from collections import defaultdict

def nested_dict(n, type):
    if n == 1:
        return defaultdict(type)
    else:
        return defaultdict(lambda: nested_dict(n-1, type))

def get_filtered_words(sentences,start_token,end_token):
    words = []
    for sentence in sentences:
      sent_words = word_tokenize(sentence)
      sent_words = start_token +  sent_words + end_token
      words += sent_words

    bad_words = ['.','#',',',"'"]
    for word in words:
      if word in bad_words:
        words.remove(word)
    return words

def calculate_word_freq(words):
    word_freq = {}
    for i in len(words):
        word_freq[words[i]] += 1
    return word_freq

# gets the pairs frequencies of chars
def get_probability(words,token_size):
    freqpairs = nested_dict(2,int)
    base_word_count = nested_dict(1,int)
    for i in range(len(words) - token_size + 1):
        token = ""
        next_token = ""
        for j in range(i,i + token_size - 1):
            token +=  words[j] + " "
            next_token += words[j+1] + " " 
        freqpairs[token][next_token] += 1
        base_word_count[token] += 1
    return freqpairs,base_word_count

def get_sentence(start_token, end_token, freqpair) :
    token = " ".join(start_token)
    token += " "
    #print(token)
    end_token = " ".join(end_token)
    end_token += " "
    #print(end_token)
    sentence = []
    i = 0
    while token != end_token:
        sentence.append(token)
        #print(token) 
        #print(freqpair[token])
        next_token = max(freqpair[token], key=freqpair[token].get)
        freqpair[token][next_token] -= 1
        token = next_token
        i += 1
    sentence.append(end_token)
    return sentence

def get_start_endline_tokens(token_size):
    start_token = []
    end_token = []
    for i in range(token_size-1):
        start_token.append("<\s>")
        end_token.append("<\e>")
    return start_token,end_token

def print_max_probibility_sentence(token_size):
    start_token,end_token = get_start_endline_tokens(token_size)
    print(start_token,  end_token)
    print()
    filtered_words = get_filtered_words(sentences,start_token,end_token)
    print(filtered_words)
    print()
    freqpair, base_freq = get_probability(filtered_words,token_size)
    print(freqpair)
    print()
    sentence = get_sentence(start_token, end_token, freqpair)
    print(sentence)
    print()
    formatted_sentence = ''
    for i in range(1,len(sentence)-token_size+1):
        split_words = sentence[i].split()
        #print(split_words)
        formatted_sentence += split_words[-1] + ' '
    print(formatted_sentence)

def main(sentences):
    token_sizes = [2,4]

    for token_size in token_sizes:
      print(token_size)
      print()
      print_max_probibility_sentence(token_size)

main(sentences)
```
