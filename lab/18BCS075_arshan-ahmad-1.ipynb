```python
import csv
import math
file = open('text.txt', 'r')
text = file.read()
```

```python
words = text.split(' ')
print('Total number of words:', len(words))
print('words:')
for w in words:
    print(w)
```

```python
vocab=set(words)
print('total vocabulary: ',len(vocab)) #number of unique words
```

```python
sentences = text.split('.')

print('total sentence: ', len(sentences))
print('sentences: ')
for s in sentences:
    print(s+".")
```

```python
sentences = set((sentences))
sentences = tuple(sentences)
print('unique sentences:')
for i in range(len(sentences)):
    print(i, sentences[i])
```
